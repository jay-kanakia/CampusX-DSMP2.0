## **<u> Abstract Classes and Methods </u>**
An **abstract class** is a class that cannot be instantiated on its own and is meant to serve as a blueprint for other classes.  
An **abstract method** is a method declared in an abstract class but contains no implementation; subclasses must provide the actual implementation.

In Python, abstract classes and methods are created using the `abc` module:

```python
from abc import ABC,abstractmethod

class BankApp(ABC):
    def database(self):
        print('connected to database')

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def display(self):
        pass
```

```python
class MobileApp(BankApp):
    def mobile_login(self):
        print('login into mobile')

    def security(self):
        print('mobile security')

    def display(self):
        print('display')
```

```python
mob = MobileApp()

mob.mobile_login()
mob.security()
mob.display()
```
