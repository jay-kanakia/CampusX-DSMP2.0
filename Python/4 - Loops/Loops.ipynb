```python
# Printing the table
number = int(input('Enter the number: '))
i = 1
while i < 11:
  print(number, '*', i, '=', number * i)
  i += 1
```

```python
# Guessing game
# generate a random integer between 1 and 100
import random
jackpot = random.randint(1, 100)

guess = int(input('Guess the number: '))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Wrong! Guess higher')
  else:
    print('Wrong! Guess lower')

  guess = int(input('Guess again: '))
  counter += 1
else:
  print('Correct guess!')
  print('Attempts:', counter)
```

```python
# Sequence sum 1/1! + 2/2! + 3/3! + .. 
n = int(input("Enter the number of terms : "))
denominator = 1
ans = 0
for i in range(1, n + 1):
  denominator *= i
  ans += i / denominator
print(ans)
```

```python
# Nested Loops - Pattern Problems - 1
rows = int(input('enter number of rows : '))
for i in range(1, rows + 1):
  for j in range(1, i + 1):
    print('*', end = '')
  print()
```

```python
# Pattern Problem - 2
rows = int(input('enter number of rows'))
for i in range(1, rows + 1):
  for j in range(1, i + 1):
    print(j, end = '')
  for j in range(i - 1, 0, -1):
    print(j, end = '')
  print()
```

```python
# Loop Control Statement
'''
- Break
- Continue
- Pass
'''
for i in range(1,10):
  if i == 5:
    break
  print(i)

lower = int(input('enter lower range'))
upper = int(input('enter upper range'))
for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)

for i in range(1,10):
  if i == 5:
    continue
  print(i)

for i in range(1,10):
  pass # This is actually used if you want to write a block of code but right now you dont know what to write in it.
```

```python
# Interview Question -> Whether the else block will executed or not
# Ans -> if a loop successifully executed till its range then the else will run but if got break in between then else will not get executed

for i in range(6):
    print(i)
    if i == 0:
        break
else:
    print("ok")
```
